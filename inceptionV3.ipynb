```python
import os
import numpy as np
import cv2
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input

# 코랩 폴더 경로
folder_path = '/content/drive/MyDrive/DatabasedAI/다리올'

# ResNet50 모델 로드
resnet_model = ResNet50(weights='imagenet', include_top=False)


# 이미지 파일명 가져오기
image_files = sorted([file for file in os.listdir(folder_path) if file.lower().endswith(('.jpg', '.jpeg', '.png'))])


# 이미지 특징 벡터를 저장할 배열 초기화
features = np.zeros((len(image_files), 7, 7, 2048))

# 특징 벡터 추출
for i, image_file in enumerate(image_files):
    img_path = os.path.join(folder_path, image_file)
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))
    image = np.expand_dims(image, axis=0)
    image = preprocess_input(image)
    features[i] = resnet_model.predict(image)

# features.npy 파일 경로
save_path = 'features.npy'

# features.npy 파일로 저장
np.save(save_path, features)

# 특징 벡터 불러오기
features = np.load(save_path)
print(features[0])
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from PIL import Image
import numpy as np
import cv2

resnet_model = ResNet50(weights='imagenet', include_top=False)

folder_path = '/content/drive/MyDrive/DatabasedAI/다리올'
img_path = '/content/drive/MyDrive/DatabasedAI/다리올/다리올01.jpg'

# 이미지 불러오기
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (224, 224))

# 이미지 전처리 및 특징 벡터 추출
x = np.expand_dims(image, axis=0)
x = preprocess_input(x)
features1 = resnet_model.predict(x)

print(features1.shape)
print(features1)
```
